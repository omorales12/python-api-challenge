```python
# Import Dependencies
import numpy as np
from citipy import citipy
import pandas as pd
from config import owkey
import requests
from pprint import pprint
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Create an empty structure for the city names
cities = []

# Loop to generate list of city names
num_countries = 500
while len(cities) < 500:
    # Generate random longitude/latitude combinations
    rand_lat = np.random.uniform(low=-90.0, high=90.0, size=None) # Latitudes range from -90 to 90
    rand_lng = np.random.uniform(low=-180.0, high=180.0, size=None) # Longitudes range from -180 to 180

    # Get the closest city to these random coordinates
    city = citipy.nearest_city(rand_lat, rand_lng)
    if city.city_name not in cities:
        cities.append(city.city_name)
```

```python
# Print first dashed line
print(50*'-')

# Initialize numpy arrays
name_arr = []
city_arr = []
country_arr = []
lat_arr = []
lng_arr = []
temp_arr = []
hum_arr = []
clo_arr = []
ws_arr = []


for city in cities:

    # Configure the URL
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&units=imperial&appid={owkey}"

    try:
        # Perform a request from the Open Weather API
        weather_data = requests.get(url).json()

        # Get parameters
        city_lat = weather_data["coord"]["lat"]
        city_lng = weather_data["coord"]["lon"]
        city_temp = weather_data["main"]["temp"]
        city_hum = weather_data["main"]["humidity"]
        city_clo = weather_data["clouds"]["all"]
        city_ws = weather_data["wind"]["speed"]
        city_name = weather_data["name"]
        
        # Append result to individual numpy arrays
        name_arr.append(city_name)
        lat_arr.append(city_lat)
        lng_arr.append(city_lng)
        temp_arr.append(city_temp)
        hum_arr.append(city_hum)
        clo_arr.append(city_clo)
        ws_arr.append(city_ws)

        # Print 
        print(f"{city_name} (lat: {city_lat}, lon:{city_lng}) has the following parameters:")
        print(f"  - Temperature: {city_temp} F")
        print(f"  - Humidity: {city_hum} %")
        print(f"  - Cloudiness: {city_clo} %")
        print(f"  - Wind Speed: {city_ws} miles/hour")
        print(50*'-')
    
    except:
        print(f'City {city} does not have complete data')
        print(50*'-')
```

```python
# Create dataframe
cities_df = pd.DataFrame({
    'City Name' : name_arr,
    'Latitude' : lat_arr,
    'Longitude' : lng_arr,
    'Temperature' : temp_arr,
    'Humidity' : hum_arr,
    'Cloudiness' : clo_arr,
    'Wind Speed' : ws_arr
})

cities_df.head()
```

```python
# Create Northern Hemisphere Dataframe
north_df = cities_df[cities_df["Latitude"] >= 0.0]
north_df.head()
```

```python
# Create Southern Hemisphere Dataframe
south_df = cities_df[cities_df["Latitude"] < 0.0]
south_df.head()
```

```python
# Save dataframe as CSV
cities_df.to_csv("weather_data.csv", encoding="utf-8", index=False)
```

```python
# Temperature (F) vs. Latitude
sns.scatterplot(x = "Latitude", y = "Temperature", data=cities_df)
sns.set_style("darkgrid")
plt.xlabel('Latitude')
plt.ylabel('Temperature (F)')
plt.title('Temperature (F) vs. Latitude')
plt.xlim(-90, 90)
plt.savefig('Images/temperature vs latitude (all).png')
plt.show()
```

```python
# Northern Hemisphere - Temperature (F) vs. Latitude
sns.scatterplot(x = "Latitude", y = "Temperature", data=north_df)
sns.set_style("darkgrid")
plt.xlabel('Latitude')
plt.ylabel('Temperature (F)')
plt.title('Northern Hemisphere - Temperature (F) vs. Latitude')
plt.xlim(0, 90)

# Linear Regression
x_values = north_df["Latitude"]
y_values = north_df["Temperature"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_values,regress_values,"r-")

# Show Plot
plt.savefig('Images/temperature vs latitude (north).png')
plt.show()
print(f"Equation: {line_eq}")
print(f"R-squared is: {rvalue**2}")
```
